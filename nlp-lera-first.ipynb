```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import pandas as pd
import re
import string
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
path_to_spam = "../input/spam-text-message-classification/SPAM text message 20170820 - Data.csv"
df = pd.read_csv(path_to_spam)
```

```python
#df.head(10)
#df.tail(10)
df
```

```python
ham = df[(df.Category == 'ham')]
spam = df[(df.Category == 'spam')]
```

```python
ham = ham['Message']
spam = spam['Message']
```

```python
hamtx = ""
spamtx = ""
for i in ham:
    hamtx += i + ' '
for i in spam:
    spamtx += i + ' '
```

```python
stop_words = stopwords.words("english")
stop_words.extend(['ok','u'])
def clean_text(text):
    text = text.lower()           #нижний
    text = re.sub("@\S+", " ", text)
    text = re.sub("https*\S+", " ", text)
    text = re.sub("#\S+", " ", text)
    text = re.sub("\d", " ", text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\s{2,}',' ', text)
    text = ' '.join([word for word in text.split(' ') if word not in stop_words])
    return text
```

```python
hamtx = clean_text(hamtx)
spamtx = clean_text(spamtx)
```

```python
hamline = word_tokenize(hamtx)
spamline = word_tokenize(spamtx) # => список слов
```

```python
lemmatizer = WordNetLemmatizer()
lemham= [lemmatizer.lemmatize(t) for t in hamline]
lemspam = [lemmatizer.lemmatize(t) for t in spamline]
```

```python
def visualize(l):
    text = ''
    for w in l:
        text += w + ' '
    wordcloud = WordCloud(width=600, height=400, background_color="green").generate(text)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
visualize(lemham)
visualize(lemspam)
```
